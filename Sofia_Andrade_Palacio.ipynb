**Algoritmos Dados**

1. **K-Nearest Neighbors (KNN) - Clasificación**
   - **Funcionamiento:** KNN es un algoritmo de clasificación supervisado que asigna la clase de una muestra nueva basándose en las clases de sus 'k' vecinos más cercanos en el espacio de características. La proximidad se mide típicamente utilizando la distancia euclidiana. El algoritmo cuenta cuántos vecinos pertenecen a cada clase y asigna la clase más común entre ellos.
   
2. **Regresión Lineal - Regresión**
   - **Funcionamiento:** La regresión lineal es un método para modelar la relación entre una variable dependiente y una o más variables independientes. El modelo busca ajustar una línea recta (en el caso de una sola variable independiente) que minimiza la suma de los errores cuadráticos entre las predicciones del modelo y los valores reales observados. La fórmula de una regresión lineal simple es \( y = mx + b \).

3. **K-Means - Clustering**
   - **Funcionamiento:** K-Means es un algoritmo de clustering no supervisado que agrupa un conjunto de datos en 'k' clusters. El algoritmo comienza seleccionando 'k' centros de cluster iniciales aleatoriamente. Luego, asigna cada punto de datos al centro más cercano y actualiza los centros de los clusters como el promedio de los puntos asignados a ellos. Este proceso se repite hasta que las asignaciones no cambien significativamente.

4. **Análisis de Componentes Principales (PCA) - Reducción de Dimensionalidad**
   - **Funcionamiento:** PCA es una técnica de reducción de dimensionalidad que transforma un conjunto de variables posiblemente correlacionadas en un conjunto de variables no correlacionadas, llamadas componentes principales. Los componentes principales se seleccionan de manera que retengan la mayor cantidad de variabilidad de los datos originales. Esto se logra mediante la descomposición de la matriz de covarianza del conjunto de datos.

### Algoritmos Adicionales

1. **Árboles de Decisión - Clasificación**
   - **Funcionamiento:** Un árbol de decisión es un modelo supervisado que divide el espacio de características en regiones con etiquetas de clase homogéneas mediante preguntas binarias. Cada nodo interno del árbol representa una característica o una prueba sobre una característica, cada rama representa el resultado de la prueba, y cada nodo hoja representa una etiqueta de clase. La construcción del árbol implica seleccionar características y umbrales que maximizan la separación de clases en cada paso, típicamente utilizando métricas como la ganancia de información o la impureza de Gini.

2. **Regresión Logística - Regresión**
   - **Funcionamiento:** La regresión logística se utiliza para modelar la probabilidad de un resultado binario basado en una o más variables independientes. Utiliza la función logística para transformar la salida lineal en una probabilidad. La fórmula es \( P(y=1|X) = \frac{1}{1 + e^{-(\beta_0 + \beta_1 X_1 + ... + \beta_n X_n)}} \), donde \( P(y=1|X) \) es la probabilidad de que el resultado sea 1 dado el conjunto de características \(X\).

3. **DBSCAN (Density-Based Spatial Clustering of Applications with Noise) - Clustering**
   - **Funcionamiento:** DBSCAN es un algoritmo de clustering basado en densidad que identifica clusters como áreas de alta densidad de puntos separados por áreas de baja densidad. A diferencia de K-Means, no requiere especificar el número de clusters con antelación. Clasifica los puntos como puntos centrales, puntos de borde o ruido, expandiendo los clusters desde los puntos centrales hasta que no se puedan agregar más puntos dentro de una distancia \( \epsilon \).

4. **t-SNE (t-distributed Stochastic Neighbor Embedding) - Reducción de Dimensionalidad**
   - **Funcionamiento:** t-SNE es una técnica de reducción de dimensionalidad no lineal que es especialmente buena para visualizar datos de alta dimensionalidad en 2 o 3 dimensiones. t-SNE minimiza la divergencia de Kullback-Leibler entre las distribuciones de puntos en el espacio de alta y baja dimensionalidad. A diferencia de PCA, t-SNE se centra en preservar la estructura local de los datos, lo que lo hace útil para la visualización de clusters complejos.

El preprocesamiento de datos es crucial para garantizar un buen rendimiento de los algoritmos de aprendizaje automático. A continuación se describen algunos pasos comunes de preprocesamiento, así como pasos específicos para los algoritmos mencionados:

### Preprocesamiento Común

1. **Limpieza de Datos:**
   - **Manejo de valores nulos:** Rellenar o eliminar valores faltantes.
   - **Eliminación de duplicados:** Identificar y eliminar registros duplicados.
   - **Corrección de errores:** Corregir errores en los datos (por ejemplo, datos mal formateados).

2. **Normalización/Estandarización:**
   - **Estandarización:** Transformar los datos para que tengan una media de 0 y una desviación estándar de 1. Es útil para algoritmos sensibles a la escala de los datos, como KNN y PCA.
   - **Normalización:** Escalar los datos a un rango fijo (por ejemplo, [0, 1]). Esto es útil para algoritmos que operan en distancias, como KNN.

3. **Codificación de variables categóricas:**
   - **One-Hot Encoding:** Transformar variables categóricas en variables binarias.
   - **Label Encoding:** Asignar un valor numérico a cada categoría.

4. **División del conjunto de datos:**
   - Dividir los datos en conjuntos de entrenamiento y prueba (y validación, si es necesario).

### Preprocesamiento Específico

1. **K-Nearest Neighbors (KNN):**
   - **Normalización/Estandarización:** KNN se basa en distancias, por lo que es importante que todas las características estén en la misma escala.
   - **Manejo de características irrelevantes:** Eliminar características irrelevantes o ruidosas, ya que pueden afectar el cálculo de distancias.

2. **Regresión Lineal:**
   - **Estandarización:** Si hay características con diferentes escalas, estandarizarlas puede ayudar a la convergencia del modelo.
   - **Eliminación de multicolinealidad:** Detectar y manejar la multicolinealidad entre las características, ya que puede afectar la interpretación del modelo y su desempeño.
   - **Manejo de valores atípicos:** Identificar y tratar los valores atípicos, ya que pueden influir significativamente en el ajuste de la línea de regresión.

3. **K-Means:**
   - **Normalización/Estandarización:** K-Means también se basa en distancias, por lo que las características deben estar en la misma escala.
   - **Selección de características:** Reducir la dimensionalidad de los datos si es necesario, utilizando técnicas como PCA antes de aplicar K-Means.
   - **Inicialización de centroides:** Usar métodos de inicialización como K-Means++ para mejorar la convergencia y la calidad de los clusters.

4. **Análisis de Componentes Principales (PCA):**
   - **Estandarización:** PCA se ve afectado por la escala de las características, por lo que es crucial estandarizarlas.
   - **Eliminación de características altamente correlacionadas:** Aunque PCA manejará la correlación, eliminar algunas características altamente correlacionadas puede simplificar el análisis.
   - **Manejo de valores atípicos:** Los valores atípicos pueden influir en los componentes principales, por lo que es importante identificarlos y tratarlos adecuadamente.

5. **Árboles de Decisión:**
   - **Manejo de valores faltantes:** Los árboles de decisión pueden manejar valores faltantes, pero puede ser útil rellenarlos o eliminarlos para simplificar el proceso.
   - **No se requiere normalización/estandarización:** Los árboles de decisión no se ven afectados por la escala de las características.

6. **Regresión Logística:**
   - **Estandarización:** Al igual que la regresión lineal, la estandarización puede ayudar en la convergencia del modelo.
   - **Manejo de multicolinealidad:** Similar a la regresión lineal, la multicolinealidad puede afectar la interpretación y el desempeño del modelo.
   - **Manejo de valores atípicos:** Los valores atípicos pueden influir en la predicción de probabilidades y la convergencia.

7. **DBSCAN:**
   - **Normalización/Estandarización:** Las distancias entre puntos son fundamentales, por lo que la normalización es crucial.
   - **Selección de características relevantes:** Elegir características relevantes puede mejorar la identificación de clusters densos.

8. **t-SNE:**
   - **Estandarización:** Aunque t-SNE es menos sensible a la escala de las características, estandarizarlas puede mejorar los resultados.
   - **Manejo de grandes conjuntos de datos:** Reducir la dimensionalidad previa (por ejemplo, con PCA) si el conjunto de datos es muy grande, ya que t-SNE puede ser computacionalmente costoso.

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import linkage, dendrogram, fcluster
import seaborn as sns
import numpy as np


file_path = '/content/Taller5_iris.csv'
iris = pd.read_csv(file_path)

# Ver las primeras filas del dataset
iris.head()
```

```python
# Separar las características y la variable objetivo
X = iris.iloc[:, :-1]
y = iris.iloc[:, -1]

# Estandarizar las características
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_scaled[:5]
```

```python
# Clustering aglomerativo
Z = linkage(X_scaled, method='ward')

# Crear el dendrograma
plt.figure(figsize=(10, 7))
dendrogram(Z)
plt.title('Dendrograma de Clustering Aglomerativo')
plt.xlabel('Índice de Muestra')
plt.ylabel('Distancia')
plt.show()
```

```python
# Elegir la altura de corte
altura_corte = 7  # Por ejemplo, podemos elegir 7 al observar el dendrograma

# Obtener los clusters
clusters_agglo = fcluster(Z, altura_corte, criterion='distance')

clusters_agglo[:5]
```

```python
# Aplicar K-Means con 3 clusters
kmeans = KMeans(n_clusters=3, random_state=42)
clusters_kmeans = kmeans.fit_predict(X_scaled)

clusters_kmeans[:5]
```

```python


# Añadir las etiquetas de clustering al dataset original
iris['Cluster_Agglo'] = clusters_agglo
iris['Cluster_KMeans'] = clusters_kmeans

# Comparar visualmente los clusters obtenidos
sns.pairplot(iris, hue='Cluster_Agglo', markers=['o', 's', 'D'])
plt.suptitle('Clusters obtenidos por Clustering Aglomerativo', y=1.02)

sns.pairplot(iris, hue='Cluster_KMeans', markers=['o', 's', 'D'])
plt.suptitle('Clusters obtenidos por K-Means', y=1.02)
plt.show()
```
